```python
import numpy as np
import load_from_snapshot
import h5py
from matplotlib import pyplot as plt
```

```python
datafolder = "/home/hlane/scpData/initTurb_stirring_0.25_0.5"
snapNum = 751
vRMSFinal = []
timeFinal = []
for i in range(snapNum): #This is for taking the FINAL Mach and Fraction#.
    ext='00'+str(i);
    if (i>=10): ext='0'+str(i)                                     #This resolves naming issues
    if (i>=100): ext=str(i)
    f = h5py.File(datafolder + "/snapshot_" + ext + ".hdf5", "r")  #opens file
    velGrad = np.array(f["PartType0"]["Velocities"])
    
    vRMSFinal.append(np.sqrt(np.sum(velGrad**2)/len(velGrad)))

    
    time = load_from_snapshot.load_from_snapshot("Time",0,datafolder,i)
    timeFinal.append(time)
```

```python
#We now move on to plotting
vRMSFinal = np.array(vRMSFinal) #array is a list of identical things, much more structured
timeFinal = np.array(timeFinal) #offers more options as lists can be anything.
plt.plot(timeFinal, vRMSFinal)   #x,y values of the lists for plotting.
plt.xlabel('time') #x label}
plt.ylabel('vRMS') #y label
```

```python
centerMassX = []
centerMassY = []
centerMassZ = []
for i in range(snapNum): #This is for taking the FINAL Mach and Fraction#.
    tenPercentList = []
    ext='00'+str(i);
    if (i>=10): ext='0'+str(i)                                     #This resolves naming issues
    if (i>=100): ext=str(i)
    f = h5py.File(datafolder + "/snapshot_" + ext + ".hdf5", "r")  #opens file
    positionsX = np.array(f["PartType0"]["Coordinates"])[:,0]
    positionsY = np.array(f["PartType0"]["Coordinates"])[:,1]
    positionsZ = np.array(f["PartType0"]["Coordinates"])[:,2]
    mass = np.array(f["PartType0"]["Masses"])
    centerMassX.append(np.sum(mass*positionsX)/np.sum(mass))
    centerMassY.append(np.sum(mass*positionsY)/np.sum(mass))
    centerMassZ.append(np.sum(mass*positionsZ)/np.sum(mass))
    
```

```python
centerMassX = np.array(centerMassX) #array is a list of identical things, much more structured
timeFinal = np.array(timeFinal) #offers more options as lists can be anything.
plt.plot(timeFinal, centerMassX)   #x,y values of the lists for plotting.
plt.xlabel('time') #x label
plt.ylabel('CenterMassX') #y label
plt.show()
centerMassY = np.array(centerMassY) #array is a list of identical things, much more structured
timeFinal = np.array(timeFinal) #offers more options as lists can be anything.
plt.plot(timeFinal, centerMassY)   #x,y values of the lists for plotting.
plt.xlabel('time') #x label
plt.ylabel('CenterMassY') #y label
plt.show()
centerMassZ = np.array(centerMassZ) #array is a list of identical things, much more structured
timeFinal = np.array(timeFinal) #offers more options as lists can be anything.
plt.plot(timeFinal, centerMassZ)   #x,y values of the lists for plotting.
plt.xlabel('time') #x label
plt.ylabel('CenterMassZ') #y label
plt.show()
```

```python
rmsDist = []
medianDist = []

for i in range(snapNum): #This is for taking the FINAL Mach and Fraction#.
    tenPercentList = []
    ext='00'+str(i);
    if (i>=10): ext='0'+str(i)                                     #This resolves naming issues
    if (i>=100): ext=str(i)
    f = h5py.File(datafolder + "/snapshot_" + ext + ".hdf5", "r")  #opens file
    distances = (f["PartType0"]["Coordinates"])
    positionsX = np.array(f["PartType0"]["Coordinates"])[:,0]
    positionsY = np.array(f["PartType0"]["Coordinates"])[:,1]
    positionsZ = np.array(f["PartType0"]["Coordinates"])[:,2]
    #print(len(np.array(f["PartType0"]["Masses"])))
    mass = np.array(f["PartType0"]["Masses"])
    cmassX = np.sum(mass*positionsX)/np.sum(mass)
    cmassY = np.sum(mass*positionsY)/np.sum(mass)
    cmassZ = np.sum(mass*positionsZ)/np.sum(mass)
    
    
    distList = []
    distX = distances[:,0]-cmassX
    distY = distances[:,1]-cmassY
    distZ = distances[:,2]-cmassZ
    distance = np.sqrt((distX**2) + (distY**2) + (distZ**2))
    #medianDist.append(np.median(distance))
    rmsDist.append(np.sqrt(np.sum((distance)**2)/len(distance)))
    dx = np.array(f["PartType0"]["Coordinates"]) - np.array([cmassX, cmassY, cmassZ]) # vector from COM
    distances = np.sqrt((dx*dx).sum(1)) # computes the distances sqrt(dX^2 + dY^2 + dZ^2)
    medianDist.append(np.median(distances))
```

```python
#We now move on to plotting
rmsDist = np.array(rmsDist) #array is a list of identical things, much more structured
timeFinal = np.array(timeFinal) #offers more options as lists can be anything.
plt.plot(timeFinal, rmsDist)   #x,y values of the lists for plotting.
plt.xlabel('time') #x label
plt.ylabel('rms Distance from Center of Mass') #y label
```

```python
medianDist = np.array(medianDist) #array is a list of identical things, much more structured
timeFinal = np.array(timeFinal) #offers more options as lists can be anything.
plt.plot(timeFinal, medianDist)   #x,y values of the lists for plotting.
plt.xlabel('time') #x label
plt.ylabel('Median Distance from Center of Mass') #y label
```
